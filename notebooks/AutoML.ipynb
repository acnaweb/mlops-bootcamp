```python
from pycaret.datasets import get_data
data1 = get_data(dataset = "traffic")
```

```python
from pycaret.regression import *
```

```python
dataset = setup(data = data1, target = "traffic_volume", session_id = 438)
```

```python
best = compare_models(exclude = ["lightgbm"])
```

```python
tuned_best = tune_model(estimator = best)
```

```python
final_model = finalize_model(estimator = tuned_best)
```

```python
model_output = "./models/traffic-pipeline"
save_model(model = final_model, model_name = "./models/traffic-pipeline")
```

### Production

```python
model = load_model(model_output)
```

```python
predictions = predict_model(model, data=data1)
```
